## Group the dataframe and sort based on the unique wounds sequentially

```python
big_df=pd.DataFrame()

sorted_df = df.sort_values('PatientIDWoundNumber')
#sorted_df=sorted_df[sorted_df.DateAdded_woundassessment.notnull()]
#sorted_df["area"]=sorted_df["area"].replace(0,0.00001)

unique_wounds=sorted_df["PatientIDWoundNumber"].unique().tolist()
for i in range(len(unique_wounds)):
    print(i)
    
    wound=unique_wounds[i]
    df_wound=sorted_df.query("PatientIDWoundNumber==@wound")
    #df_wound=df_wound[["PatientIDWoundNumber","DateAdded_woundassessment","area"]]
    df_wound=df_wound.drop_duplicates()
    df_wound=df_wound[df_wound.area.notnull()]

    
    # Convert 'DateAdded_woundassessment' column to datetime type
    #df_wound['DateAdded_woundassessment'] = pd.to_datetime(df_wound['DateAdded_woundassessment'])

    # Keep the maximum value in 'area' for each date
    df_wound = df_wound.loc[df_wound.groupby('DayNumber')['area'].idxmax()]

    df_wound=df_wound.sort_values('DayNumber')
    
    big_df = pd.concat([big_df, df_wound], axis=0)
    
```

```python
# Finding the most number of visits across all patients and wounds
grouped_data = big_df.groupby(['PatientIDWoundNumber'])['DayNumber'].count()
maximum = grouped_data.max()

print(grouped_data)
print("MAX:" + str(maximum))
```

## Atika's..

```python
transposed_big_df=pd.DataFrame()

list_unique_wound=big_df["PatientIDWoundNumber"].unique().tolist()

# Columns to transpose
# list_to_transpose
# list_not_transpose

for i in range(len(list_unique_wound)):
#     if i>40:
#         break
    print(i)
    unique_wound_number=list_unique_wound[i]
    wound_df=big_df.query("PatientIDWoundNumber== @unique_wound_number")
    #display(wound_df)
    
    df=wound_df
    # Transpose selected columns
    transposed_df = pd.concat([df[list_to_transpose].shift(-i).add_suffix(f'_Visit{i+1}') for i in range(len(df))], axis=1)

    # Concatenate transposed columns with non-transposed columns
    result_df = pd.concat([df[list_not_transpose], transposed_df], axis=1)
    
    result_df=result_df.iloc[:1]
    #display(result_df)
    
    wound_columns = [col for col in result_df.columns if col.startswith('Wound_healing_status_Visit')]

    # Extract the column with the highest number after "Visit"
    max_visit_number = max(int(col.split('Visit')[1]) for col in wound_columns)
    max_visit_column = f'Wound_healing_status_Visit{max_visit_number}'
    print("\n",max_visit_column)
    # Extract the values from the selected column
    result_df['FinalVisit_Healing_Status'] = result_df[max_visit_column]
    
    # Specify the columns to be moved to the front
    columns_to_front = ['PatientIDWoundNumber', 'FinalVisit_Healing_Status']

    # Move the specified columns to the front
    result_df = pd.concat([result_df[columns_to_front], result_df.drop(columns=columns_to_front)], axis=1)

    #display(result_df)
    
    transposed_big_df=pd.concat([transposed_big_df, result_df], axis=0)
    
```
